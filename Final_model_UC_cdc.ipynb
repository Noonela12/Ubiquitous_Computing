```python
# Remove the 'Target' column from the DataFrame
df = df.drop('Target', axis=1)

def new_labeling_scheme(row):
    if row['Battery_percentage'] > 70:
        return 0  # High battery, prioritize low power consumption
    elif 45 <= row['Battery_percentage'] <= 70:
        power_accuracy_tradeoff = [row['GPS_power'], row['Wifi_power'], row['Cellular_power']]
        min_power = min(power_accuracy_tradeoff)
        if min_power == row['GPS_power']:
            return 1
        elif min_power == row['Wifi_power']:
            return 1
        else:
            return 1
    elif 25 <= row['Battery_percentage'] < 45:
        return 2  # Low battery, minimize power consumption
    else:
        return 3  # Critical battery, extreme power saving

# Apply the new labeling scheme
df['Target'] = df.apply(new_labeling_scheme, axis=1)

# Now retrain the model and evaluate as before
```
